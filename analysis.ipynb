```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import analysis_tools as tools
```

```python
FILE = "result.json"
COLUMN_ORDER = ["mistral:7b", "gemma:7b", "qwen2.5:7b", "llama3.1:8b", "dolphin3:8b", "deepseek-r1:8b"]
ROW_ORDER = ["cycle_detection", "subgraph_isomorphism", "planarity", "graph_coloring", "traversal_count"]
```

```python
df0, df1 = tools.load_data(FILE)
```

### Zero-shot

```python
print("Zero-shot: Counts")
count_table_0 = tools.gen_count_table(df0, COLUMN_ORDER, ROW_ORDER)

plt.figure(figsize=(2, 2))
sns.heatmap(count_table_0, annot=True, cmap="Blues", fmt="d", linewidths=0.5)
plt.title("Counts Heatmap by Model and Task")
plt.ylabel("Task")
plt.xlabel("Model")
plt.yticks(rotation=0)
plt.show()
```

```python
print("Zero-shot: Accuracy per model per task (%)")
accuracy_table_0 = tools.gen_accuracy_table(df0, COLUMN_ORDER, ROW_ORDER)

plt.figure(figsize=(9, 2))
sns.heatmap(accuracy_table_0, annot=True, cmap="viridis", fmt=".1f", linewidths=0.5)
plt.title("Accuracy Heatmap by Model and Task")
plt.ylabel("Task")
plt.xlabel("Model")
plt.yticks(rotation=0)
plt.show()
```

```python
print("Zero-shot: Accuracy per model (%)")
accuracy_model_table_0 = tools.gen_accuracy_model_table(df0, COLUMN_ORDER, ROW_ORDER)

plt.figure(figsize=(1, 2))
sns.heatmap(accuracy_model_table_0, annot=True, cmap="viridis", fmt=".1f", linewidths=0.5)
plt.title("Accuracy Heatmap by Model")
plt.ylabel("Model")
plt.yticks(rotation=0)
plt.xticks([])
plt.show()
```

```python
print("Zero-shot: Accuracy per task (%)")
accuracy_task_table_0 = tools.gen_accuracy_task_table(df0, COLUMN_ORDER, ROW_ORDER)

plt.figure(figsize=(1, 2))
sns.heatmap(accuracy_task_table_0, annot=True, cmap="viridis", fmt=".1f", linewidths=0.5)
plt.title("Accuracy Heatmap by Task")
plt.ylabel("Task")
plt.yticks(rotation=0)
plt.xticks([])
plt.show()
```

```python
print("Zero-shot: Accuracy per model per prompt (boolean)")
accuracy_task_tables_0 = tools.gen_accuracy_task_tables(df0, ROW_ORDER, COLUMN_ORDER)

for task, accuracy_task_table in accuracy_task_tables_0.items():
    print(f"'{task}':")

    plt.figure(figsize=(10, 2))
    sns.heatmap(accuracy_task_table, annot=True, cmap="viridis", fmt="d", linewidths=0.5)
    plt.ylabel("Model")
    plt.xlabel("Prompt")
    plt.yticks(rotation=0)
    plt.show()
```

```python
print("Zero-shot: Average response time per model per task (sec)")
time_table_0 = tools.gen_time_table(df0, COLUMN_ORDER, ROW_ORDER)

plt.figure(figsize=(10, 2))
sns.heatmap(time_table_0, annot=True, cmap="viridis", fmt=".1f", linewidths=0.5)
plt.title("Average Time Heatmap by Model and Task")
plt.ylabel("Task")
plt.xlabel("Model")
plt.yticks(rotation=0)
plt.show()
```
